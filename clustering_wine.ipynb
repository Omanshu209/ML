```python
df_main = df[["Alcohol", "Color_Intensity"]]
```

```python
plt.scatter(df_main.iloc[:, 0], df_main.iloc[:, 1])
plt.xlabel("Alcohol")
plt.ylabel("Color_Intensity")
```

```python
from sklearn.cluster import KMeans
```

```python
sse = []

for k in range(1, 11):
  model = KMeans(n_clusters = k)
  model.fit(df_main.values)
  sse.append(model.inertia_)
```

## **Elbow Method**

```python
plt.plot(sse)
plt.xticks(labels = [i for i in range(1, 11)], ticks = [i for i in range(10)])
plt.xlabel("n_clusters")
plt.ylabel("SSE")
```

**`n_clusters` = 3 (or 4)**

```python
model = KMeans(n_clusters = 3)
df_main["cluster"] = model.fit_predict(df_main.values)
```

```python
df_main.cluster.unique()
```

```python
centers = model.cluster_centers_
centers
```

```python
df_0 = df_main[df_main.cluster == 0]
df_1 = df_main[df_main.cluster == 1]
df_2 = df_main[df_main.cluster == 2]

plt.scatter(df_0.iloc[:, 0], df_0.iloc[:, 1], color = "blue", label = '0')
plt.scatter(df_1.iloc[:, 0], df_1.iloc[:, 1], color = "green", label = '1')
plt.scatter(df_2.iloc[:, 0], df_2.iloc[:, 1], color = "red", label = '2')

# Plotting The Centers Of The Clusters
plt.scatter(centers[:, 0], centers[:, 1], color = "black", marker = '*')

plt.xlabel("Alcohol")
plt.ylabel("Color_Intensity")
plt.legend()
```
