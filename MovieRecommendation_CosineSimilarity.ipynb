# **Movie Recommendation System - Cosine Similarity**

**cos(θ) = (A • B) / ( |A| |B| ),**

**where θ is the angle between vectors A and B**

```python
genres_set = set({})

for item in df.genres:
  genres = item.split("},")
  for genre in genres:
    i = genre.split("name\": ")
    if len(i) < 2:
      continue
    if i[1][1 : -1].endswith("}"):
      continue
    genres_set.add(i[1][1 : -1])

genres_set
```

```python
import math
```

```python
df_new["magnitude"] = pd.Series(dtype = float)
```

```python
df_new.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

for col in df_new.columns[: 7]:
  df_new[col] = df_new[col].astype(float)
  df_new[col] = scaler.fit_transform(pd.DataFrame(df_new[col]))
```

```python
df_new.head(7)
```

```python
for i in range(4800):
  sumSquare = 0
  for col in df_new.columns:
    if col == "magnitude":
      continue
    try:
      element = df_new[col][i]
      sumSquare += element * element
    except Exception:
      pass
  df_new["magnitude"][i] = math.sqrt(sumSquare)
```

```python
df_new.head()
```

```python
np_arr = df_new.values
np_arr[0]
```

```python
np_arr.shape
```

```python
def get_movie_index(movie_name : str) -> int:
  i = 0
  while df["title"][i].upper() != movie_name.upper():
    if i == 4800:
      return -1
    i += 1
  return i
```

```python
def get_cosines(movie_name : str):
  index = get_movie_index(movie_name)
  cosines = np.array([])

  i = 0
  while i < 4800:
    cosines = np.append(
        cosines,
        [np.dot(np_arr[i][: -1], np_arr[index][: -1]) / (np_arr[i][-1] * np_arr[index][-1])],
        axis = 0
    )
    i += 1

  return cosines
```

```python
get_cosines("frozen")
```

```python
def suggest(movie_name : str, n : int):
  cosines = get_cosines(movie_name)
  cosines_sorted = np.sort(cosines)
  movies = []

  for i in range(1, n + 1):
    index = 0
    while(cosines_sorted[-i] != cosines[index] and index < 4799):
      index += 1
    movies.append(df["title"][index])

  return movies
```

```python
suggest("toy story", 10)[3:]
```
